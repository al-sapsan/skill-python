## Вариант 1: Решето Эратосфена

### Описание алгоритма:
Решето Эратосфена — это классический алгоритм для поиска всех простых чисел до \( n \). Алгоритм начинается с создания списка чисел от 2 до \( n \) и последующего исключения всех кратных каждого найденного простого числа.

### Код:


```python
def eratosthenes_sieve(max_num):
    sieve = [True] * (max_num + 1)
    sieve[0] = sieve[1] = False  # 0 и 1 не являются простыми числами
    
    for i in range(2, int(max_num**0.5) + 1):
        if sieve[i]:
            for j in range(i * i, max_num + 1, i):
                sieve[j] = False
    
    primes = [i for i in range(2, max_num + 1) if sieve[i]]
    return primes

# Вывод чисел по десяткам
def print_primes(primes):
    for i in range(0, len(primes), 10):
        print(", ".join(map(str, primes[i:i+10])) + ",")

max_num = int(input("Введите максимальное число: "))
print("Простые числа:")
primes = eratosthenes_sieve(max_num)
print_primes(primes)
```

### Скорость:
Алгоритм работает за время \( O(n \log(\log(n))) \), где \( n \) — максимальное число. Это один из самых быстрых алгоритмов для нахождения всех простых чисел в заданном диапазоне.

### Преимущества:
- Эффективен для больших диапазонов чисел.
- Простота реализации и понимания.

### Операторы:
- `sieve[i] = False`: исключает числа, которые не являются простыми.
- `range(i * i, max_num + 1, i)`: выбирает кратные числа для исключения.

## Вариант 2: Прямой метод проверки делителей

### Описание алгоритма:
Этот алгоритм проверяет каждое число от 2 до \( n \), является ли оно простым, проверяя делимость этого числа на все простые числа, меньшие его квадратного корня. Если число делится без остатка на любое из этих чисел, оно не является простым.

### Код:


```python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

def find_primes(max_num):
    primes = []
    for i in range(2, max_num + 1):
        if is_prime(i):
            primes.append(i)
    return primes

# Вывод чисел по десяткам
def print_primes(primes):
    for i in range(0, len(primes), 10):
        print(", ".join(map(str, primes[i:i+10])) + ",")

max_num = int(input("Введите максимальное число: "))
print("Простые числа:")
primes = find_primes(max_num)
print_primes(primes)
```

### Скорость:
Алгоритм работает за время \( O(n \sqrt{n}) \), где \( n \) — максимальное число. Это значительно медленнее, чем Решето Эратосфена, особенно при больших значениях \( n \).

### Преимущества:
- Легко понять и реализовать.
- Подходит для небольших диапазонов чисел.

### Операторы:
- `num % i == 0`: проверка, делится ли число на другой делитель без остатка.
- `int(num**0.5) + 1`: определяет границу поиска делителей.

## Вариант 3: Оптимизированное Решето Эратосфена (использование только нечётных чисел)

### Описание алгоритма:
Этот алгоритм является улучшенной версией Решета Эратосфена, где проверяются только нечётные числа. Число 2 добавляется в список простых сразу, а затем применяются те же шаги, что и в первом варианте, но только для нечётных чисел.

### Код:


```python
def optimized_sieve(max_num):
    if max_num < 2:
        return []
    
    sieve = [True] * ((max_num // 2) + 1)
    sieve[0] = False  # 1 не является простым числом
    
    for i in range(1, int(max_num**0.5) // 2 + 1):
        if sieve[i]:
            for j in range(2 * i * (i + 1), len(sieve), 2 * i + 1):
                sieve[j] = False
    
    primes = [2] + [2 * i + 1 for i in range(1, len(sieve)) if sieve[i]]
    return primes

# Вывод чисел по десяткам
def print_primes(primes):
    for i in range(0, len(primes), 10):
        print(", ".join(map(str, primes[i:i+10])) + ",")

max_num = int(input("Введите максимальное число: "))
print("Простые числа:")
primes = optimized_sieve(max_num)
print_primes(primes)
```

### Скорость:
Как и у обычного Решета Эратосфена, время работы — \( O(n \log(\log(n))) \), но благодаря работе только с нечётными числами он может быть в два раза быстрее в среднем.

### Преимущества:
- Более эффективное использование памяти.
- Быстрее для больших диапазонов чисел, чем классическое Решето Эратосфена.

### Операторы:
- `sieve[0] = False`: устанавливает, что 1 не является простым.
- `[2 * i + 1 for i in range(1, len(sieve)) if sieve[i]]`: генерирует список простых чисел.

## Заключение

- **Решето Эратосфена** — наилучший выбор для задач с большими диапазонами чисел, благодаря его высокой эффективности.
- **Прямой метод проверки делителей** — простой в понимании, но подходит для небольших диапазонов из-за своей относительной медлительности.
- **Оптимизированное Решето Эратосфена** — улучшенная версия, которая экономит время и память, работая только с нечётными числами.
