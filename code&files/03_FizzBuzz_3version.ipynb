# Задача 3. FizzBuzz 

### Вариант1:

1. **Функция `fizz_buzz(max_number)`**:
   - Проверяет, что `max_number` (максимальное число) больше или равно 1. Если нет, выводит сообщение об ошибке.
   - Проходит по всем числам от 1 до `max_number` включительно.
   - Для каждого числа проверяет:
     - Если число делится на 3 и 5 одновременно, выводит "fizz buzz".
     - Если число делится только на 3, выводит "fizz".
     - Если число делится только на 5, выводит "buzz".
     - Если число не делится ни на 3, ни на 5, выводит само число.

2. **Функция `main()`**:
   - Запрашивает у пользователя ввод максимального числа.
   - Пытается преобразовать ввод в целое число и передать его в `fizz_buzz`.
   - Если ввод не является целым числом, выводит сообщение об ошибке.

3. **Запуск программы**:
   - Если код выполняется как основной скрипт, вызывается функция `main()`, которая запускает основной процесс.

```python
def fizz_buzz(max_number):
    if max_number < 1:
        print("Максимальное число должно быть не меньше 1")
        return
    
    for i in range(1, max_number + 1):
        if i % 3 == 0 and i % 5 == 0:
            print("fizz buzz")
        elif i % 3 == 0:
            print("fizz")
        elif i % 5 == 0:
            print("buzz")
        else:
            print(i)

def main():
    try:
        max_number = int(input("Введите максимальное число: "))
        fizz_buzz(max_number)
    except ValueError:
        print("Пожалуйста, введите целое число")

if __name__ == "__main__":
    main()
```

## Вариант 2  
оптимизированная версия кода, которая более компактна и использует более эффективные подходы:

### Изменения и улучшения:
1. **Список с выражением**:
   - Вместо использования цикла `for` и последовательных проверок, применяется список с выражением (list comprehension). Это сокращает код и делает его более читаемым.

2. **Возврат значений**:
   - Функция `fizz_buzz` теперь возвращает либо список, либо строку с сообщением об ошибке. Это делает ее более универсальной, и можно легко использовать результаты функции в других частях программы без необходимости сразу выводить их.

3. **Проверка результата**:
   - В функции `main()` проверяется, является ли результат строкой (ошибочное сообщение), или списком (результат FizzBuzz), и соответствующим образом выводится на экран.

4. **Конкатенация строк**:
   - Использование `"\n".join(result)` позволяет вывести результат сразу в виде одного многострочного блока, что упрощает код.

```python
def fizz_buzz(max_number):
    if max_number < 1:
        return "Максимальное число должно быть не меньше 1"
    
    return [
        "fizz buzz" if i % 3 == 0 and i % 5 == 0 else
        "fizz" if i % 3 == 0 else
        "buzz" if i % 5 == 0 else
        str(i)
        for i in range(1, max_number + 1)
    ]

def main():
    try:
        max_number = int(input("Введите максимальное число: "))
        result = fizz_buzz(max_number)
        if isinstance(result, str):
            print(result)
        else:
            print("\n".join(result))
    except ValueError:
        print("Пожалуйста, введите целое число")

if __name__ == "__main__":
    main()
```

## Вариант2. Улучшенный

Для улучшения читаемости и оптимизации предложенного кода, можно внести несколько изменений:

### Изменения:
1. **Улучшение структуры условий**: Заменим вложенные тернарные операторы на более читаемый `if-elif-else` блок внутри функции `fizz_buzz`. Это улучшит читаемость и понимание логики.

2. **Изменение подхода к обработке ошибок**: Вместо возврата строки ошибки из `fizz_buzz`, перенесем проверку и сообщение об ошибке в функцию `main`. Это отделяет логику расчета от логики обработки ошибок и делает код более модульным.

3. **Рефакторинг вывода результатов**: Упростим проверку результата в `main`, так как ошибка теперь обрабатывается до вызова `fizz_buzz`.

### Оптимизированный и улучшенный код:

```python
def fizz_buzz(max_number):
    fizz_buzz_list = []
    for i in range(1, max_number + 1):
        if i % 3 == 0 and i % 5 == 0:
            fizz_buzz_list.append("fizz buzz")
        elif i % 3 == 0:
            fizz_buzz_list.append("fizz")
        elif i % 5 == 0:
            fizz_buzz_list.append("buzz")
        else:
            fizz_buzz_list.append(str(i))
    return fizz_buzz_list

def main():
    try:
        max_number = int(input("Введите максимальное число: "))
        if max_number < 1:
            print("Максимальное число должно быть не меньше 1")
            return
        
        result = fizz_buzz(max_number)
        print("\n".join(result))
    except ValueError:
        print("Пожалуйста, введите целое число")

if __name__ == "__main__":
    main()
```


### Примененные изменения:

1. **`if-elif-else` вместо тернарных операторов**:
   - Внутри функции `fizz_buzz` использован более читаемый и стандартный блок `if-elif-else`, который легко воспринимается и не требует тщательного анализа вложенности тернарных операторов.
   - Это делает код более структурированным и понятным даже для начинающих программистов.

2. **Обработка ошибок до вызова `fizz_buzz`**:
   - Проверка на невалидное значение `max_number` (`< 1`) перемещена в `main`. Теперь функция `fizz_buzz` отвечает только за вычисление списка, а не за обработку ошибок.
   - Такой подход делает функции более узконаправленными, что улучшает их тестируемость и переиспользуемость.

3. **Прямой вывод результата**:
   - Если входное число корректное, результат сразу выводится, без необходимости проверки типа результата. Это уменьшает количество проверок и делает код проще.

4. **Обработка ввода и ошибок**:
   - Проверка на корректность ввода (`try-except`) осталась в `main`, что позволяет пользователю ввести данные заново в случае ошибки.
   - Это также делает код более устойчивым и удобным для пользователя.

Эти изменения улучшают читаемость, структуру и модульность кода, делая его более понятным и легким в сопровождении.

## Вариант 3
еще один вариант оптимизированного кода, который использует функциональный подход и еще более компактное решение:


### Изменения и улучшения:
1. **Использование булевых выражений**:
   - Внутри списка с выражением используются булевы выражения (`i % 3 == 0` и `i % 5 == 0`), которые умножаются на строки. Если условие истинно, строка включается в результат, иначе возвращается пустая строка.

2. **Упрощение логики**:
   - Условная логика объединена в одно выражение с помощью цепочки `or`. Это устраняет необходимость в использовании вложенных условных операторов `if-else` и делает код более линейным и легко читаемым.

3. **Чёткое сообщение об ошибке**:
   - В `main()` при вводе отрицательного числа или нуля выводится сообщение об ошибке, без вызова `fizz_buzz`.

4. **Сокращение кода**:
   - Код стал короче и проще, сохранив всю функциональность.

```python
def fizz_buzz(n):
    return ["fizz buzz" * (i % 3 == 0 and i % 5 == 0) or
            "fizz" * (i % 3 == 0) or
            "buzz" * (i % 5 == 0) or
            str(i)
            for i in range(1, n + 1)]

def main():
    try:
        max_number = int(input("Введите максимальное число: "))
        if max_number < 1:
            print("Максимальное число должно быть не меньше 1")
        else:
            print("\n".join(fizz_buzz(max_number)))
    except ValueError:
        print("Пожалуйста, введите целое число")

if __name__ == "__main__":
    main()
```
